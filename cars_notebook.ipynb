## Data Cleaning

```python
# Creating df_clean as cleaned version from the original dataset (df)
df_clean = df.copy()

# Imputing null values in Customer Name Column with 'Unknown'
df_clean['Customer Name'] = df_clean['Customer Name'].fillna('Unknown')
# Changing datatype in Date column to datetime
df_clean['Date'] = pd.to_datetime(df_clean['Date'])
# Changing datetype in Phone column to categorical (string)
df_clean['Phone'] = df_clean['Phone'].astype('str')
# Creating 'Month_Year column from Date column
df_clean['Month_Year'] = df_clean['Date'].dt.to_period('m')
# Changing Month_Year column data type to string
df_clean['Month_Year'] = df_clean['Month_Year'].astype('str')
# Reordering column
df_clean = df_clean[['Car_id','Date','Month_Year','Dealer_Region','Dealer_Name','Company','Model','Body Style','Color','Engine','Transmission','Price ($)','Customer Name','Gender','Annual Income','Phone']]

df_clean
```

### Dealing with outlier

```python
# Ammual Income
q3_income = df_clean['Annual Income'].quantile(0.75)
q1_income = df_clean['Annual Income'].quantile(0.25)
t_income = q3_income - q1_income
income_outlier_above = q3_income+1*t_income
income_outlier_below = q1_income-1*t_income


# Price
q3_price= df_clean['Price ($)'].quantile(0.75)
q1_price = df_clean['Price ($)'].quantile(0.25)
t_price = q3_price - q1_price
price_outlier_above = q3_price+1.5*t_price
price_outlier_below = q1_price-1.5*t_price

print(f'Upper limit in Annual Income column :{income_outlier_above}\nLower limit in Annual Income column :{income_outlier_below}\nUpper limit in Price column : {price_outlier_above}\nLower limit in Price column : {price_outlier_below}')
df_clean = df_clean[(df_clean['Annual Income'] > income_outlier_below) & (df_clean['Annual Income'] < income_outlier_above)]
df_clean = df_clean[(df_clean['Price ($)'] > price_outlier_below) & (df_clean['Price ($)'] < price_outlier_above)]

df_clean
```

### Dealer Performance

```python
## Store Performance 2022-2023
store_perf = df_clean.groupby('Dealer_Name')['Car_id'].count().sort_values(ascending=False).reset_index()
store_perf = store_perf.rename(columns={'Car_id':'Sales Made'})
store_perf
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(x=store_perf['Sales Made'],y=store_perf['Dealer_Name'],orient='h',palette='rainbow')
plt.title('Numbers of Sales Made from Each Dealer')
```

```python
# Revenue
rev_store = df_clean.groupby('Dealer_Name')['Price ($)'].sum().sort_values(ascending=False).reset_index()
rev_store
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(x=rev_store['Price ($)'],y=rev_store['Dealer_Name'],orient='h',palette='rainbow')
plt.title('Total Revenue from Each Dealer')
plt.xlabel('Revenue ($)')
```

Classic Chevy and Enterprise Rent A Car are the worst performer among all of the dealers, meanwhile Iceberg Rentals has the lowest car sales, but the revenue are much more better.

##### Revenue by total car sold

```python
# Joining total revenue and total car sold from each dealer
store_eff = pd.merge(rev_store,store_perf,how='inner',on='Dealer_Name')
#Renaming price to total revenue
store_eff = store_eff.rename(columns={'Price ($)':'Revenue ($)'})
# Add new column which divide total_revenue with sales made
store_eff['Revenue per Unit Sold'] = (store_eff['Revenue ($)']/store_eff['Sales Made']).round(2)


store_eff = store_eff.sort_values('Revenue per Unit Sold',ascending=False)

store_eff
```

```python
plt.subplots(figsize=(20,10))

sns.barplot(x=store_eff['Revenue per Unit Sold'],y=store_eff['Dealer_Name'],orient='h',palette='rainbow')
plt.title('Revenue per Unit Sold in Each Dealer')
```
